### 1. Write a program that asks for three numbers and prints the largest one

```python
num_1 = int(input("Enter the first number:"))

num_2 = int(input("Enter the second number:"))

num_3 = int(input("Enter the third number:"))

largest = max(num_1, num_2, num_3)

print(f"The Largest number is : {largest}")
```

### 2. Create a function that takes a string as input and prints each character on a new line.

```python
greeting =input("Please enter a greeting")
print(f"Each character are : ")
for letter in greeting:
    print(letter)
```

### 3. Implement a program to calculate the sum of all even numbers between 1 and 100.

```python
total = sum(range(2, 101, 2))
print(total)
```

### 4. Create a function that checks if a number is positive, negative, or zero.

```python
num = int(input("Input a number"))
if num >0:
    print("The number is Positive")
elif num < 0:
    print("The number is Negative")
else:
    print("The number is Zero")
```
