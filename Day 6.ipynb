## 1. Create a function that uses a loop to find the GCD of two numbers. 

```python
def find_gcd(a, b):
    gcd = 1  # start with the smallest possible GCD
    
    for i in range(1, min(a, b) + 1):
        if a % i == 0 and b % i == 0:
            gcd = i  # update gcd if i divides both numbers
            
    return gcd
```

```python
num1 = int(input("Enter first number: "))
num2 = int(input("Enter second number: "))

gcd = find_gcd(num1, num2)
print(f"The GCD of {num1} and {num2} is {gcd}")
```

### 4.Create a function that prints all prime numbers between 1 and 100. 


```python
def print_primes():
    for num in range(2, 101):
        count = 0

        for i in range(1, num + 1):
            if num % i == 0:
                count += 1

        if count == 2:
            print(num, end=" ")
```

```python
print_primes()
```

### 5. Write a program that uses nested loops to find all pairs of numbers in a list that add up to a given target.

```python
numbers = [1,2, 4, 3, 5, 7, 8, 9]
target = 9

pairs = []

for i in range(len(numbers)):
    for j in range(i + 1, len(numbers)):
        if numbers[i] + numbers[j] == target:
            pairs.append((numbers[i], numbers[j]))

print(pairs)
```
