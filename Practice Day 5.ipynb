### 2. Create a function to calculate the sum of numbers in a list until a number greater than 50 is found. 

```python
def sum_list(numbers):
    total = 0
    
    for num in numbers:
        if num > 50:
            break
        total += num
    
    return total
```

```python
my_list= [1,35,4,6,4,7,10,9,20,13]
sum_list(my_list)
```

```python
my_list= [1,35,4,60,4,7,10,9,20,13]
sum_list(my_list)
```

### 3.  Implement a nested loop to create a pattern (e.g., a right-angled triangle of stars).


```python
for i in range(1,6):
    for j in range(i,6):
        print(" ", end = " ")
    for k in range (0,i):
        print("*", end = " ")
    for m in range (i , 1, -1):
        print("*", end = " ")
    print(" ")
```

### 4. Write a program to count how many numbers between 1 and 100 are divisible by both 3 and 5. 

```python
count = 0

for num in range(1, 101):
    if num % 3 == 0 and num % 5 == 0:
        count += 1

print("Count:", count)
```

### 5.  Use a while loop to print the Fibonacci sequence up to a user-provided limit.

```python
# Get user input
user_input = int(input("Enter the number limit: "))

# First two  number of Fibonacci numbers
a = 0
b = 1

# Print the sequence using while loop
while a <= user_input:
    print(a, end=" ")
    a, b = b, a + b
```
