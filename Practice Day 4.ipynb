### 1. Write a program to print all the numbers from 1 to 20 using a for loop.

### 2.Create a function that prints "Hello" 10 times using a while loop.

```python
def print_hello():
    count = 0  # start counter
    
    while count < 10:  # loop until count reaches 10
        print("Hello")
        count += 1  # increase counter by 1

```

```python
# Call the function


print_hello()
```

### 3.Use an if-else statement to check if a given number is greater than 10. 

```python
num = int(input("Enter any number"))

if num > 10:
    print(num, "is greater than 10")
elif num < 10:
    print(num, "is not greater than 10")
else:
    print(num, "is equal to 10")
```

### 4. Create a program that prints the numbers 1 to 10 but skips 5 using a continue statement.

```python
for number in range(1, 11):  # numbers from 1 to 10
    
    if number == 5:
        
        continue  # skip the rest of this loop when number is 5
        
    print(number)
```

### 5. Write a function to find the sum of all integers from 1 to a user-provided number. 


```python
def sum_integers(number):
    total = 0
    for i in range(1, number + 1):
        total += i
    return total

# Ask the user for input
number = int(input("Enter a positive integer: "))
print("The sum of integers from 1 to", number, "is:", sum_integers(number))
```
