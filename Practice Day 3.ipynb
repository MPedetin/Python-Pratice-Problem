### 1.  Implement a function that finds the factorial of a number using a loop
.

```python
num = int(input("Enter a number:"))

factorial = 1

for i in range(1, num+1):
    factorial = factorial * i
    
print("The factorial of ", num, "is" ,factorial)
```

### OR

```python
num = int(input("Enter a number:"))

factorial = 1 

if num < 0 :
    print("The factorial does not exist")
    
elif num == 0:
    print("The factorial is 1")
    
else:
    for n in range(1, num+1):
        factorial *=n

        
print("The factorial of", num, "is", factorial)
```

### 2. Write a function to calculate the average of a list of numbers provided by the user.

```python

num1 = int(input("Emter num1:"))

num2 = int(input("Emter num2:"))

num3 = int(input("Emter num3:"))

num4 = int(input("Emter num4:"))

numbers = [num1, num2, num3, num4]
 
average = sum(numbers) / len(numbers)

print("The average of", numbers ,"is" , average )
```

### 3. Create a program to reverse a given string.

```python
word = input("Enter a word")

print("The reverse of", word, "is",   word[::-1])
```

### 5. Write a function to determine if a number is prime.

```python
# allow user enter a number

num = int(input("Enter a number")) 

prime = 0   # a counter to count how many number divides num starting from zero

for  x in range (1, num +1):
    if (num % x == 0):
        prime += 1
        
if prime ==2:
    print(num , " is a prime number")
else :
    print(num, "is not a prime number")
        
```
